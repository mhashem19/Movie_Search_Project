```python
import warnings
warnings.simplefilter('ignore')

# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

movies_df = pd.read_csv('data/imdb_movies.csv')
movies_df.head(3)
```

```python
# This is commented because we ran this block of code successfully
# Removing $, EUR, AD currency indicators and turn columns into Integer from Object.
movies_df[movies_df.columns[10:12]] = movies_df[movies_df.columns[10:12]].replace('[$,EUR,AD,AUD,ER]', '', regex=True).astype(int)
```

```python
movies_df
```

```python
movies_df = movies_df[["duration", "genre", "avg_vote", "budget", "usa_gross_income"]].dropna()
```

```python
movies_df
```

```python
movies_df["genre"].value_counts()
```

```python
expanded = pd.concat([movies_df[["duration", "avg_vote", "budget", "usa_gross_income"]], movies_df["genre"].str.split(", ",expand=True)],axis=1)
```

```python
expanded = expanded [["duration", "avg_vote", "budget", "usa_gross_income", 0]]
```

```python
expanded
```

```python
expanded = expanded.rename(columns={0:"genre"})
expanded 
```

```python
dummies = pd.get_dummies(expanded["genre"])
```

```python
expanded[dummies.columns] = dummies
del expanded["genre"]
expanded
```
